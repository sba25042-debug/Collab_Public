```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
```

```python
df = pd.read_csv("apartments_for_rent_classified_semi_colon_sep.csv", sep=";", encoding='latin1')
```

```python
df.dropna(inplace=True)
```

```python
X = df[["square_feet", "bathrooms", "bedrooms", "longitude", "latitude"]]
y = df["price"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=72
)
```

```python
random_state = 77
rf_model = RandomForestRegressor(n_estimators=100, random_state=random_state)
rf_model.fit(X_train, y_train)

dt_model = DecisionTreeRegressor(random_state=random_state)
dt_model.fit(X_train, y_train)
```

```python
rf_y_pred = rf_model.predict(X_test)
dt_y_pred = dt_model.predict(X_test)

print("Random Forest MSE:", mean_squared_error(y_test, rf_y_pred))
print("Random Forest MAPE:", mean_absolute_percentage_error(y_test, rf_y_pred))
print("Random Forest R2:", r2_score(y_test, rf_y_pred))

print("*"*50)

print("Decision Tree MSE:", mean_squared_error(y_test, dt_y_pred))
print("Decision Tree MAPE:", mean_absolute_percentage_error(y_test, dt_y_pred))
print("Decision Tree R2:", r2_score(y_test, dt_y_pred))
```
